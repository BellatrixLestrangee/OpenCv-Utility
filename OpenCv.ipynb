```
img = cv2.imread('jp.png')
print(type(img))
```

```
h,w,d = img.shape
print("Height = {}, Width = {}, Depth={}".format(h,w,d))
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(img)
cv2.waitKey(0)
```

```
#Checking RGB value for a pixel coordinate
(r, g, b) = img[100][20]   #red, green ,blue
print("Blue = {} Green = {} Red = {}".format(b,g,r))
```

```
#Resizing Image using OpenCV
resized_img = cv2.resize(img, (200, 200))
cv2_imshow(resized_img)
cv2.waitKey(0)
```

```
#Calculating and Adjusting the Aspect Ratio
res_width = 200/w 
res_h = res_width* h
resized_img = cv2.resize(img, (200, int(res_h)))
cv2_imshow(resized_img)
cv2.waitKey(0)
```

```
#Imutil Solves the problem of aspect ratio
resized_img = imutils.resize(img, width = 200)
cv2_imshow(resized_img)
cv2.waitKey(0)
```

```
#parameter in imutil can be adjusted as height or width both
resized_img = imutils.resize(img, height = 200)
cv2_imshow(resized_img)
cv2.waitKey(0)
```

```
#Rotation using WarpAffine and CV2
center = (w//2, h//2)
rotation_Matrix = cv2.getRotationMatrix2D(center, -45, 1)
rotated = cv2.warpAffine(img, rotation_Matrix, (w, h))
cv2_imshow(rotated)
cv2.waitKey(0)
```

```
#Rotation using Imutils
rotated = imutils.rotate(img, 45)
cv2_imshow(rotated)
cv2.waitKey(0)
```

```
# To see the whole picture, we gotta upgrade
rotated = imutils.rotate_bound(img, -45)
cv2_imshow(rotated)
cv2.waitKey(0)
```

```
 #Drawing a rectangle 
 copy_img = img.copy()   #we use a copy of image to leave the original image unharmed
 # parameters include (image, left upper corner, right bottom corner, BGR color values for line, thickness of line)
 cv2.rectangle(copy_img, (320,60), (420, 160), (0,0,255), 2)   
 cv2_imshow(copy_img)
 cv2.waitKey(0)
```

```
#drawing a circle
copy_img_2 = img.copy()
cv2.circle(copy_img_2, (100,100), 30, (255, 0, 0), 2)
cv2_imshow(copy_img_2)
cv2.waitKey(0)
```

```
copy_img_3 = img.copy()
cv2.line(copy_img_3, (50,50), (200, 50), (0, 255, 0), 2)
cv2_imshow(copy_img_3)
cv2.waitKey(0)
```

```
copy_img_4 = img.copy()
cv2.putText(copy_img_4,"Heyo! Tutorial to CV2", (30,30), cv2.FONT_HERSHEY_DUPLEX, 0.5, (123,255,0), 2 )
cv2_imshow(copy_img_4)
cv2.waitKey(0)
```
