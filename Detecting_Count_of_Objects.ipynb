```
import numpy
import cv2
import imutils
import os
from google.colab.patches import cv2_imshow
```

```
path = "/content/drive/My Drive/opencv-tutorial"
os.chdir(path)
os.listdir('.')
```

```
img = cv2.imread('sample01.png')
cv2_imshow(img)
cv2.waitKey(0)
```

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
cv2.waitKey(0)
```

```
#canny is the edge detecting algorithm
edged = cv2.Canny(gray, 40, 150)
cv2_imshow(edged)
cv2.waitKey(0)
```

```
thresh = cv2.threshold(gray, 225,255, cv2.THRESH_BINARY_INV)[1]
cv2_imshow(thresh)
cv2.waitKey(0)
```

```
gray.shape
```

```
cont = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cont = imutils.grab_contours(cont)
output = img.copy()

for c in cont:
  cv2.drawContours(output, [c],-1, (0,0,255), 3)
  
cv2_imshow(output)
cv2.waitKey(0)
```

```
text = "{} objects found".format(len(cont))
cv2.putText(output, text, (10,25), cv2.FONT_HERSHEY_PLAIN, 2, 2)
cv2_imshow(output)
cv2.waitKey(0)
```
